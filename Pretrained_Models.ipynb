```python
from tensorflow.keras.applications.resnet50 import ResNet50
```

```python
from tensorflow.keras.preprocessing import image
```

```python
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
```

```python
import numpy as np
```

```python
model = ResNet50(weights='imagenet')  ## imagenet dataset is trained on huge data
```

```python
img_path = '/content/lambo.png'
img = image.load_img(img_path, target_size=(224, 224))
x = image.img_to_array(img)
x = np.expand_dims(x, axis=0)
x = preprocess_input(x)  ## it will do all the preprocessing of the image
```

```python
preds = model.predict(x)
```

```python
print('Predicted:', decode_predictions(preds, top=3)[0]) ## give all three possible categories
```
